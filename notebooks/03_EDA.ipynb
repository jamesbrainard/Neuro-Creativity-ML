```python
# ==============================
# REQUIRED LIBRARIES
# ==============================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# ==============================
# LOADING DATA
# ==============================
DATA_DIR = "../data/features"
df = pd.read_csv(os.path.join(DATA_DIR, "data.csv"))
```

```python
# ==============================
# VISUALIZING BRAIN FEATURE DISTRIBUTIONS
# ==============================
brain_feature_cols = [col for col in df.columns if any(k in col for k in ["fa", "conn", "activation"])]

for col in brain_feature_cols[250:256]:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Distribution of {col}")
    plt.tight_layout()
    plt.show()
```
