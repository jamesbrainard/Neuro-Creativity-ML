```python
# ==============================
# REQUIRED LIBRARIES
# ==============================
import os
import nibabel as nib
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
# ==============================
# LOADING DATA FROM 'participants.tsv'
# ==============================
DATA_DIR = "../data/raw"
participants_data_path = os.path.join(DATA_DIR, "participants.tsv")
df_participants = pd.read_csv(participants_data_path, sep="\t")
```

```python
for col in df_participants:
    plt.figure(figsize=(3, 2))
    plt.hist(df_participants[col].dropna(), bins = 20, edgecolor = 'black', alpha = 0.7)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
# ==============================
# LOADING AND PLOTTING A TW1-WEIGHTED MRI SCAN
# ==============================

# Defines the path to the dataset
DATA_DIR = "../data/raw"
SAMPLE_SUBJECT = "sub-01"

# Constructs the full path to the file & loads image
t1w_path = os.path.join(DATA_DIR, SAMPLE_SUBJECT, "anat", f"{SAMPLE_SUBJECT}_T1w.nii.gz")
t1_img = nib.load(t1w_path)

# Prints the image shape, which shows the pixel count for X, Y, and Z values.
print(f"T1-weighted image shape: {t1_img.shape}")

# Converts image to numpy array
t1_data = t1_img.get_fdata()

# Displays a middle slice of the image
plt.imshow(t1_data[:, :, t1_data.shape[2] // 2], cmap="gray")
plt.title(f"T1-weighted Image - {SAMPLE_SUBJECT}")
plt.axis("off")
plt.show()
```

```python
# ==============================
# LOADING AND PLOTTING A SAMPLE FMRI SCAN
# ==============================
# Defines the path to the dataset
DATA_DIR = "../data/raw"
SAMPLE_SUBJECT = "sub-01"

# Constructs the full path to the file & loads image
fmri_path = os.path.join(DATA_DIR, SAMPLE_SUBJECT, "func", f"{SAMPLE_SUBJECT}_task-rest_run-01_bold.nii.gz")
fmri_img = nib.load(fmri_path)

# Prints the image shape, which shows the pixel count for X, Y, Z, and time-series values.
print(f"fMRI Image Shape: {fmri_img.shape}")

# Converts image to numpy array
fmri_data = fmri_img.get_fdata()
timepoint = fmri_data[:, :, fmri_data.shape[2] // 2, 10] # Gets 10th timepoint just for this image

# Displays the image at the given timepoint
plt.imshow(timepoint, cmap="hot")
plt.title(f"fMRI Slice at Timepoint 10 - {SAMPLE_SUBJECT}")
plt.axis("off")
plt.show()
```
